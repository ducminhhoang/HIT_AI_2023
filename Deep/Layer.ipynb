# Layer
## Khái niệm
- Về bản chất, DL bao gồm nhiều layers, mỗi layer có thể coi là một mạng Neural Network (NN). Mỗi hidden layer được gọi là fully connected layer, tên gọi theo đúng ý nghĩa, mỗi node trong hidden layer được kết nối với tất cả các node trong layer trước. Cả mô hình được gọi là fully connected neural network (FCN)
## Tại sao phải sử dụng Convolution NN?
Hạn chế của các phương pháp trước là:
- Chỉ chạy tốt với những hình ảnh chỉ có đối tượng đó, và đối tượng nằm ở giữa
- Sử dụng nhiều parameters

Bên cạnh hiệu năng cao trên số lượng mẫu cần thiết để đạt được đủ độ chính xác, các mạng nơ-ron tích chập thường có hiệu quả tính toán hơn, bởi đòi hỏi ít tham số hơn và dễ thực thi song song trên nhiều GPU

## Convolutional layer

Trong quá trình covolution: output feature map phụ thuộc kích thước của input feature map, kích thước của kernel, và stride.
> stride: khoảng cách giữa 2 lần trượt (default: stride = 1)
![hahha](https://stanford.edu/~shervine/teaching/cs-230/illustrations/parameter-compatibility-en.jpeg?bc91caf0473dc42f1a495946f67726d3)

- Convolutional (tích chập): Convolutional layer (CONV) sử dụng filter (bộ lọc) để thực hiện các tính toán chập  giúp giảm đáng kể số lượng tham số cần học so với các mô hình fully connected và giúp lấy ra các đặc trưng của dữ liệu đầu vào.
- **filter (kernel/ feature detector):** là matrix có kích thước nhỏ được sử dụng lấy ra các thông tin chính xác nhất mà không cần phải chọn các feature.

- Ouput Feature Map = ma trận (tích vô hướng của ma trận vùng input chiếu đang xét với ma trận filter)

![](https://stanford.edu/~shervine/teaching/cs-230/illustrations/convolution-layer-a.png?1c517e00cb8d709baf32fc3d39ebae67)

## Padding layer
- Padding (đệm): là việc thêm các giá trị 0 vào xung quanh ma trận đầu vào trước khi convolutional. Mục đích để giữ nguyên kích thước của ma trận đầu ra so với ma trận đầu vào.
- Có 2 loại padding phổ biến: valid padding và same padding, full padding


## Pooling layer
- Pooling (gộp): là một downsampling operations, thường được dùng sau khi convolutional layer trong mạng CNN
- Mục đích: Tóm tắt các features đã được tích chập 
=> Làm giảm kích thước của features map
### Max pooling
- Mục đích: Mỗi 1 operation sẽ chọn giá trị max trong các ô đang xét
- Nhận xét:
> Được sử dụng nhiều nhất

> Bảo tồn được các features
![](https://stanford.edu/~shervine/teaching/cs-230/illustrations/max-pooling-a.png?711b14799d07f9306864695e2713ae07)
### Average pooling
- Mục đích: Mỗi 1 operation sẽ lấy giá trị trung bình của các ô đang xét
- Nhận xét:
> Downsamples feature map

> Được sử dụng trong LeNet
![](https://stanford.edu/~shervine/teaching/cs-230/illustrations/average-pooling-a.png?58f9ab6d61248c3ec8d526ef65763d2f)
